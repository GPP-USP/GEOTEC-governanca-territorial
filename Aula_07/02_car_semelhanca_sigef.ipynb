# <span style="color:#000000">4. Cálculo do Indicador: Semelhança CAR & SIGEF </span>
<hr style="border:1px solid #000000;">


<p>
Nesta etapa, iremos calcular o indicador de semelhança entre um imóvel do Cadastro Ambiental Rural (CAR) e os registros do Sistema de Gestão Fundiária (SIGEF).
<p>

<p>
Considera-se que um imóvel do CAR apresenta semelhança com o SIGEF quando há alta coincidência entre as áreas.
Em termos práticos, isso ocorre quando pelo menos <b>90%</b> da área do CAR coincide com o SIGEF e, simultaneamente,
<b>90%</b> da área do SIGEF coincide com o CAR.
</p>

<br>

<p>
O indicador verifica essa correspondência espacial por meio de duas proporções:
</p>

<div style="background-color:#f8f9fa; padding:15px; border-radius:8px; text-align:center; font-size:1.2em;">
<b>p₁ = Área(CAR ∩ SIGEF) / Área(CAR)</b><br>
<b>p₂ = Área(CAR ∩ SIGEF) / Área(SIGEF)</b>
</div>

<p>
As duas proporções são necessárias: enquanto p₁ indica quanto do CAR está coberto pelo SIGEF,
p₂ indica quanto do SIGEF está contido no CAR.
</p>

<p>
<b>Critério de semelhança:</b> considera-se que os imóveis são semelhantes quando
<b>p₁ &gt; 0,9</b> e <b>p₂ &gt; 0,9</b>, ou seja, quando há alta correspondência entre as áreas em ambas as direções.
</p>


<br>
<p>
<b>O limiar de 90% pode variar conforme critério adotado.​</b>
</p>

<br>

###  4.2.1: Selecionar e visualizar o CAR de Interesse
Vamos localizar o imóvel no banco de dados do CAR e visualizar sua geometria.

#### Seleciona e visualiza as informações do CAR

```python
# ============================================================================
# Consulta SQL para selecionar o imóvel do CAR
# ============================================================================

query_car = f"""
SELECT *
FROM {CAR_SCHEMA}.{CAR_TABLE}
WHERE {CAR_ID} = '{COD_IMOVEL}'
"""
# ============================================================================
# Exibição da query 
# ============================================================================
md_query = f"""
<div style="background-color:#f5f5f5; padding:15px; border-radius:8px; border-left: 5px solid #2c6e9e; font-family: monospace; font-size: 0.9em;">
<b> Consulta SQL executada:</b><br>
<code>
{query_car.replace('<', '&lt;').replace('>', '&gt;')}
</code>
</div>
"""

display(Markdown(md_query))
```

#### Executa a consulta

```python
# ============================================================================
# Carrega o imóvel como GeoDataFrame
# ============================================================================
gdf_car = gpd.read_postgis(query_car, engine, geom_col=CAR_GEOM)

# ============================================================================
# Visualização das informações do Imóvel selecionado (versão simplificada)
# ============================================================================

md_resultado = f"""
<div style="background-color:#e8f4fd; padding:15px; border-radius:8px; border-left: 5px solid #2196f3; margin: 15px 0;">
<b>Imóvel selecionado no CAR:</b><br>
Código: <code>{COD_IMOVEL}</code><br>
</div>
"""

display(Markdown(md_resultado))

# Exibe o DataFrame completo para você ver todas as colunas disponíveis
display(gdf_car.head())
```

#### Visualização da geometria CAR


```python
# -----------------------------------------------------------------------------
# Verifica se o imóvel foi carregado corretamente na etapa anterior
# -----------------------------------------------------------------------------
if 'gdf_car' not in locals() or gdf_car.empty:
    print("X Imóvel do CAR não encontrado. Execute a célula 'Selecionar o CAR de Interesse' primeiro.")
    
    # Cria um mapa padrão centralizado no Brasil como fallback
    map_center_location = [-15.7801, -47.9292]
    m = folium.Map(location=map_center_location, zoom_start=4, control_scale=True, tiles='Esri.WorldImagery')
    display(m)

else:
    # -------------------------------------------------------------------------
    # Preparar a geometria para visualização
    # -------------------------------------------------------------------------
    # O Folium funciona com coordenadas geográficas (latitude/longitude)
    # O CRS original do CAR já é SIRGAS 2000 (EPSG:4674), que é geográfico
    # Portanto, podemos usar diretamente sem reprojetar!
    gdf_car_for_display = gdf_car

    # -------------------------------------------------------------------------
    # Converter colunas do tipo data/hora para texto
    # O Folium converte os dados para JSON antes de enviar para o navegador
    # O JSON não suporta objetos datetime do Python
    # -------------------------------------------------------------------------
    gdf_car_json_safe = gdf_car_for_display.copy()
    for col in gdf_car_json_safe.columns:
        if pd.api.types.is_datetime64_any_dtype(gdf_car_json_safe[col]):
            gdf_car_json_safe[col] = gdf_car_json_safe[col].astype(str)

    # -------------------------------------------------------------------------
    # Calcular o centro e os bounds do polígono (sem gerar warning)
    # -------------------------------------------------------------------------
    # Usando total_bounds que é mais eficiente e não gera warning
    bounds = gdf_car.geometry.total_bounds  # [minx, miny, maxx, maxy]
    
    # Calcula o centro a partir dos bounds (sem precisar do centroid)
    map_center_location = [(bounds[1] + bounds[3]) / 2, (bounds[0] + bounds[2]) / 2]

    # -------------------------------------------------------------------------
    # Criar o mapa interativo
    # -------------------------------------------------------------------------
    m = folium.Map(
        location=map_center_location, 
        zoom_start=10,  # valor temporário, será substituído pelo fit_bounds
        control_scale=True, 
        tiles='Esri.WorldImagery',
        width='60%',      # largura da célula
        height=300        # Altura fixa em pixels
    )

    # -------------------------------------------------------------------------
    # Ajustar o zoom exato para o polígono
    # -------------------------------------------------------------------------
    # fit_bounds ajusta o mapa para enquadrar perfeitamente o polígono
    m.fit_bounds([[bounds[1], bounds[0]], [bounds[3], bounds[2]]])

    # -------------------------------------------------------------------------
    # Adicionar o polígono do CAR ao mapa
    # Cor branca conforme padrão definido
    # -------------------------------------------------------------------------
    folium.GeoJson(
        gdf_car_json_safe,
        name="CAR",
        style_function=lambda x: {
            'color': 'white',           # Contorno branco
            'weight': 3,                # Espessura da borda
            'fillColor': 'white',       # Preenchimento branco
            'fillOpacity': 0.4          # Transparência de 40%
        },
        tooltip=f"CAR: {COD_IMOVEL}"
    ).add_to(m)

    # -------------------------------------------------------------------------
    # Exibir o mapa
    # -------------------------------------------------------------------------
    display(m)

    # ============================================================================
    # Libera memória do mapa
    # ============================================================================
    if 'm' in locals():
        del m
        gc.collect()
```

### 4.2.2 Identificar e visualizar as parcelas do SIGEF que intersectam o imóvel selecionado

Nesta etapa, vamos identificar e visualizar as parcelas do SIGEF que apresentam interseção com o imóvel do CAR selecionado, preparando os dados para o cálculo das áreas de sobreposição.

#### Identifica as parcelas do SIGEF

```python
# Consulta SQL para encontrar parcelas do SIGEF que intersectam o CAR
query_sigef = f"""
WITH car_geom AS (
    SELECT {CAR_GEOM} AS geom
    FROM {CAR_SCHEMA}.{CAR_TABLE}
    WHERE {CAR_ID} = '{COD_IMOVEL}'
)
SELECT s.*
FROM {SIGEF_SCHEMA}.{SIGEF_TABLE} s
INNER JOIN car_geom c
    ON ST_Intersects(s.{SIGEF_GEOM}, c.geom)

"""

# ============================================================================
# Exibição da query 
# ============================================================================
md_query_sigef = f"""
<div style="background-color:#f5f5f5; padding:15px; border-radius:8px; border-left: 5px solid #2c6e9e; font-family: monospace; font-size: 0.9em;">
<b> Consulta SQL executada:</b><br>
<code>
{query_sigef.replace('<', '&lt;').replace('>', '&gt;')}
</code>
</div>
"""

display(Markdown(md_query_sigef))
```

```python
# ============================================================================
# Carrega as parcelas como GeoDataFrame
# ============================================================================
gdf_sigef = gpd.read_postgis(query_sigef, engine, geom_col=SIGEF_GEOM)

# ============================================================================
# Exibe o resultado
# ============================================================================
display(Markdown(f"#### Parcelas SIGEF encontradas: **{len(gdf_sigef)}**"))

display(gdf_sigef)
```

### 4.2.3: Calcular a área de intersecção entre as entre CAR e SIGEF

Vamos calcular a área de intersecção entre o CAR e as parcelas do SIGEF.

 </b><br> Todas as áreas serão calculadas utilizando a projeção Albers (SRID 97823) para garantir consistência e precisão nos cálculos, conforme definido na metodologia do projeto. </div>

#### Calcular a área de intersecção entre o CAR e as parcelas

```python
query_calculo_interseccao = f"""
WITH car_geom AS (
    SELECT {CAR_GEOM} AS geom
    FROM {CAR_SCHEMA}.{CAR_TABLE}
    WHERE {CAR_ID} = '{COD_IMOVEL}'
),
sigef_intersect AS (
    SELECT
        s.nome_area,
        s.parcela_co,
        s.codigo_imo,
        ST_Area(ST_Transform(c.geom, 97823)) / 10000 AS area_car,
        ST_Area(ST_Transform(s.{SIGEF_GEOM}, 97823)) / 10000 AS area_sigef_ha,
        ST_Area(ST_Transform(ST_Intersection(s.{SIGEF_GEOM}, c.geom), 97823)) / 10000 AS area_intersecao_ha
    FROM {SIGEF_SCHEMA}.{SIGEF_TABLE} s
    INNER JOIN car_geom c
    ON ST_Intersects(s.{SIGEF_GEOM}, c.geom)
)
SELECT
    * from sigef_intersect;
"""

# ============================================================================
# Exibição da query 
# ============================================================================
md_query_calculo_interseccao = f"""
<div style="background-color:#f5f5f5; padding:15px; border-radius:8px; border-left: 5px solid #2c6e9e; font-family: monospace; font-size: 0.9em;">
<b> Consulta SQL executada:</b><br>
<code>
{query_calculo_interseccao.replace('<', '&lt;').replace('>', '&gt;')}
</code>
</div>
"""

display(Markdown(md_query_calculo_interseccao))
```

```python
# ============================================================================
# Executa a consulta e carrega os resultados em um DataFrame
# ============================================================================
df_areas = pd.read_sql(text(query_calculo_interseccao), con=engine)

# ============================================================================
# Exibe o resultado
# ============================================================================
display(Markdown(f"#### Detalhamento por Parcela: **{len(gdf_sigef)}**"))

display(df_areas)
```

###  4.2.5 Indicador de Semelhança CAR & SIGEF

```python
output = f"""
WITH car_geom AS (
    SELECT {CAR_GEOM} AS geom
    FROM {CAR_SCHEMA}.{CAR_TABLE}
    WHERE {CAR_ID} = '{COD_IMOVEL}'
),
sigef_intersect AS (
    SELECT
        s.nome_area,
        s.parcela_co,
        s.codigo_imo,
        ST_Area(ST_Transform(c.geom, 97823)) / 10000 AS area_car,
        ST_Area(ST_Transform(s.{SIGEF_GEOM}, 97823)) / 10000 AS area_sigef_ha,
        ST_Area(ST_Transform(ST_Intersection(s.{SIGEF_GEOM}, c.geom), 97823)) / 10000 AS area_intersecao_ha
    FROM {SIGEF_SCHEMA}.{SIGEF_TABLE} s
    INNER JOIN car_geom c
    ON ST_Intersects(s.{SIGEF_GEOM}, c.geom)
),
calculo_proporcoes_parcela AS (
    SELECT
        'por_parcela' AS tipo,
        a.codigo_imo,
        a.parcela_co,
        a.nome_area,
        a.area_car,
        a.area_sigef_ha,
        a.area_intersecao_ha,
        a.area_intersecao_ha / a.area_car AS p1,
        a.area_intersecao_ha / a.area_sigef_ha AS p2
    FROM sigef_intersect a
),
calculo_proporcoes_agrupamento AS (
    SELECT
        'Agrupamento' AS tipo,
        a.codigo_imo,
        'Agrupamento' AS parcela_co,
        'Agrupamento' AS nome_area,
        max(a.area_car) AS area_car,
        sum(a.area_sigef_ha) AS area_sigef_ha,
        sum(a.area_intersecao_ha) AS area_intersecao_ha,
        sum(a.area_intersecao_ha) / max(a.area_car) AS p1,
        sum(a.area_intersecao_ha) / sum(a.area_sigef_ha) AS p2
    FROM sigef_intersect a
    GROUP BY a.codigo_imo
),
indicador_final AS (
SELECT *,
       CASE
           WHEN p1 > {limiar} AND p2 > {limiar}  THEN TRUE
           ELSE FALSE
       END AS semelhante_sigef
FROM calculo_proporcoes_parcela

UNION ALL

SELECT *,
       CASE
           WHEN p1 > {limiar} AND p2 > {limiar}  THEN TRUE
           ELSE FALSE
       END AS semelhante_sigef
FROM calculo_proporcoes_agrupamento)
select
  '{COD_IMOVEL}' as cod_car,
  codigo_imo,
  tipo,
  semelhante_sigef
from  indicador_final
where semelhante_sigef
GROUP BY codigo_imo, semelhante_sigef, tipo
ORDER BY codigo_imo;
;
"""

# ============================================================================
# Exibição da query 
# ============================================================================
md_output = f"""
<div style="background-color:#f5f5f5; padding:15px; border-radius:8px; border-left: 5px solid #2c6e9e; font-family: monospace; font-size: 0.9em;">
<b> Consulta SQL executada:</b><br>
<code>
{output.replace('<', '&lt;').replace('>', '&gt;')}
</code>
</div>
"""

display(Markdown(md_output))
```

| Critério | Interpretação |
|----------|---------------|
| p₁ > 0,9 e p₂ > 0,9 | ✔️ **CAR é semelhante ao SIGEF**|
| p₁ ≤ 0,9 ou p₂ ≤ 0,9 | ⚠️ **CAR não é semelhante ao SIGEF** |
