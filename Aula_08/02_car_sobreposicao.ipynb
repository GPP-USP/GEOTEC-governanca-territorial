# <span style="color:#000000">4. Cálculo do Indicador: Sobreposição entre CARs </span>
<hr style="border:1px solid #000000;">

<p>
Nesta etapa, iremos analisar as sobreposições espaciais entre polígonos do Cadastro Ambiental Rural (CAR).
</p>

<p>
A existência de interseção entre dois ou mais cadastros gera incerteza sobre a delimitação declarada do imóvel e pode indicar conflitos potenciais de limites, afetando a segurança quanto aos direitos associados à terra.
</p>

<br>

<p>
O indicador está estruturado em duas dimensões complementares: <b>intensidade</b> e <b>qualidade da sobreposição</b>, operacionalizadas por três métricas.
</p>



### Intensidade da sobreposição

A intensidade mede o grau de sobreposição de um imóvel do CAR com os demais imóveis da base, considerando dois aspectos:

<ul>
  <li><b>(Métrica A) — Número de CARs sobrepostos:</b> quantidade de imóveis que apresentam interseção com o CAR de interesse;</li>
  <li><b>(Métrica B) — Percentual de área sobreposta:</b> proporção da área do CAR que está sobreposta a outros imóveis.</li>
</ul>

<p>
Essa dimensão permite avaliar o quanto um imóvel está espacialmente comprometido por sobreposições.
</p>


### Qualidade da sobreposição

A qualidade da sobreposição busca qualificar essas interseções,para isso, avalia-se se há sobreposição com algum CAR que apresente <b>semelhança com o SIGEF (Métrica C)</b>, conforme o indicador definido anteriormente.

<br>

###  4.2.1: Selecionar e visualizar o CAR de Interesse
Vamos localizar o imóvel no banco de dados do CAR e visualizar sua geometria.



#### Seleciona e visualiza as informações do CAR

```python
# ============================================================================
# Consulta SQL para selecionar o imóvel do CAR
# ============================================================================

query_car = f"""
SELECT DISTINCT ON ({CAR_ID}) * 
FROM {CAR_SCHEMA}.{CAR_TABLE}
WHERE {CAR_ID} = '{COD_IMOVEL}'
    AND {CAR_STATUS} <> 'CA'
    AND {CAR_CONDICAO} NOT ILIKE '%cancelado%'
    ORDER BY {CAR_ID}, COALESCE({CAR_DATA_ATUALIZACAO}, {CAR_DATA_CRIACAO}) DESC;
"""


# ============================================================================
# Exibição da query 
# ============================================================================
md_query = f"""
<div style="background-color:#f5f5f5; padding:15px; border-radius:8px; border-left: 5px solid #2c6e9e; font-family: monospace; font-size: 0.9em;">
<b> Consulta SQL executada:</b><br>
<code>
{query_car.replace('<', '&lt;').replace('>', '&gt;')}
</code>
</div>
"""

display(Markdown(md_query))
```

#### Executa a consulta

```python
# ============================================================================
# Carrega o imóvel como GeoDataFrame
# ============================================================================
gdf_car = gpd.read_postgis(text(query_car), engine, geom_col=CAR_GEOM)

# ============================================================================
# Visualização das informações do Imóvel selecionado (versão simplificada)
# ============================================================================

md_resultado = f"""
<div style="background-color:#e8f4fd; padding:15px; border-radius:8px; border-left: 5px solid #2196f3; margin: 15px 0;">
<b>Imóvel selecionado no CAR:</b><br>
Código: <code>{COD_IMOVEL}</code><br>
</div>
"""

display(Markdown(md_resultado))

# Exibe o DataFrame completo para você ver todas as colunas disponíveis
display(gdf_car.head())
```

#### Visualização da geometria CAR


```python
# -----------------------------------------------------------------------------
# Verifica se o imóvel foi carregado corretamente na etapa anterior
# -----------------------------------------------------------------------------
if 'gdf_car' not in locals() or gdf_car.empty:
    print("X Imóvel do CAR não encontrado. Execute a célula 'Selecionar o CAR de Interesse' primeiro.")
    
    # Cria um mapa padrão centralizado no Brasil como fallback
    map_center_location = [-15.7801, -47.9292]
    m = folium.Map(location=map_center_location, zoom_start=4, control_scale=True, tiles='Esri.WorldImagery')
    display(m)

else:
    # -------------------------------------------------------------------------
    # Extrai informações do imóvel para o tooltip
    # -------------------------------------------------------------------------
    cod_imovel = gdf_car['cod_imovel'].iloc[0]
    status = gdf_car['ind_status'].iloc[0]
    municipio = gdf_car['municipio'].iloc[0]
    uf = gdf_car['cod_estado'].iloc[0]

    # -------------------------------------------------------------------------
    # Preparar a geometria para visualização
    # -------------------------------------------------------------------------
    # O Folium funciona com coordenadas geográficas (latitude/longitude)
    # O CRS original do CAR já é SIRGAS 2000 (EPSG:4674), que é geográfico
    # Portanto, podemos usar diretamente sem reprojetar!
    gdf_car_for_display = gdf_car

    # -------------------------------------------------------------------------
    # Converter colunas do tipo data/hora para texto
    # O Folium converte os dados para JSON antes de enviar para o navegador
    # O JSON não suporta objetos datetime do Python
    # -------------------------------------------------------------------------
    gdf_car_json_safe = gdf_car_for_display.copy()
    for col in gdf_car_json_safe.columns:
        if pd.api.types.is_datetime64_any_dtype(gdf_car_json_safe[col]):
            gdf_car_json_safe[col] = gdf_car_json_safe[col].astype(str)

    # -------------------------------------------------------------------------
    # Calcular o centro e os bounds do polígono (sem gerar warning)
    # -------------------------------------------------------------------------
    # Usando total_bounds que é mais eficiente e não gera warning
    bounds = gdf_car.geometry.total_bounds  # [minx, miny, maxx, maxy]
    
    # Calcula o centro a partir dos bounds (sem precisar do centroid)
    map_center_location = [(bounds[1] + bounds[3]) / 2, (bounds[0] + bounds[2]) / 2]

    # -------------------------------------------------------------------------
    # Criar o mapa interativo
    # -------------------------------------------------------------------------
    m = folium.Map(
        location=map_center_location, 
        zoom_start=10,  # valor temporário, será substituído pelo fit_bounds
        control_scale=True, 
        tiles='Esri.WorldImagery',
        width='60%',      # largura da célula
        height=300        # Altura fixa em pixels
    )

    # -------------------------------------------------------------------------
    # Ajustar o zoom exato para o polígono
    # -------------------------------------------------------------------------
    # fit_bounds ajusta o mapa para enquadrar perfeitamente o polígono
    m.fit_bounds([[bounds[1], bounds[0]], [bounds[3], bounds[2]]])

    # -------------------------------------------------------------------------
    # Criar o tooltip com as informações do imóvel
    # -------------------------------------------------------------------------
    tooltip_text = f"""
    <div style="font-family: Arial, sans-serif;">
        <b>CAR</b><br>
        <b>Código:</b> {cod_imovel}<br>
        <b>Status:</b> {status}<br>
        <b>Município:</b> {municipio} / {uf}
    </div>
    """
    
    # -------------------------------------------------------------------------
    # Adicionar o polígono do CAR ao mapa
    # Cor branca conforme padrão definido
    # -------------------------------------------------------------------------
    folium.GeoJson(
        gdf_car_json_safe,
        name="CAR",
        style_function=lambda x: {
            'color': 'white',           # Contorno branco
            'weight': 3,                # Espessura da borda
            'fillColor': 'white',       # Preenchimento branco
            'fillOpacity': 0.4          # Transparência de 40%
        },
        tooltip=folium.Tooltip(tooltip_text)
    ).add_to(m)

    # -------------------------------------------------------------------------
    # Exibir o mapa
    # -------------------------------------------------------------------------
    display(m)

    # ============================================================================
    # Libera memória do mapa
    # ============================================================================
    if 'm' in locals():
        del m
        gc.collect()
```

### 4.2.4: Calcular o percentual de sobreposição (Métrica B)

Nesta etapa, vamos calcular o **percentual de área do imóvel do CAR que está sobreposta a outros imóveis**, correspondente à Métrica B do indicador.

Esse percentual é obtido a partir da razão entre:
- a área total de interseção entre o CAR de interesse e os demais imóveis;
- e a área total do imóvel do CAR analisado.

Formalmente, o cálculo pode ser representado por:

$$
\text{Perc. de sobreposição} = \frac{\text{Área sobreposta}}{\text{Área do CAR}} \times 100
$$

Onde:
- **Área sobreposta** corresponde à soma das áreas de interseção entre o CAR de interesse e os demais imóveis identificados;
- **Área do CAR** corresponde à área total do imóvel analisado.



#### Calculo do percentual de sobreposição

```python
# ============================================================================
# Métrica B - Percentual de sobreposição 
# ============================================================================

query_perc_area_sobreposta_total = f"""
WITH car_interesse AS (
    SELECT DISTINCT ON ({CAR_ID})
        {CAR_ID},
        {CAR_GEOM} AS geom
    FROM {CAR_SCHEMA}.{CAR_TABLE}
    WHERE {CAR_ID} = '{COD_IMOVEL}'
        AND {CAR_STATUS} <> 'CA'
        AND {CAR_CONDICAO} NOT ILIKE '%cancelado%'
    ORDER BY 
        {CAR_ID}, 
        COALESCE({CAR_DATA_ATUALIZACAO}, {CAR_DATA_CRIACAO}) DESC
),
car_sobrepostos AS (
    SELECT 
        b.*
    FROM car_interesse a
    INNER JOIN {CAR_SCHEMA}.{CAR_TABLE} b
        ON ST_Intersects(a.geom, b.{CAR_GEOM})
        AND ST_Area(ST_Transform(ST_Intersection(a.geom, b.{CAR_GEOM}), 97823)) > 0
    WHERE 
        b.{CAR_ID} <> a.{CAR_ID}
        AND b.{CAR_STATUS} <> 'CA'
        AND b.{CAR_CONDICAO} NOT ILIKE '%cancelado%'
),
car_sobrepostos_unicos AS (
    SELECT DISTINCT ON ({CAR_ID})
        *
    FROM car_sobrepostos
    ORDER BY 
        {CAR_ID},
        COALESCE({CAR_DATA_ATUALIZACAO}, {CAR_DATA_CRIACAO}) DESC
)
SELECT
    -- área do imóvel em hectares
    ST_Area(ST_Transform(a.geom, 97823)) / 10000 AS area_imovel_ha,
    -- área total sobreposta em hectares (união de todos os CARs sobrepostos)
    ST_Area(ST_Transform(ST_Intersection(a.geom, ST_Union(b.{CAR_GEOM})), 97823)) / 10000 AS area_sobreposta_ha,
    -- percentual da área do imóvel que está sobreposta
    (ST_Area(ST_Transform(ST_Intersection(a.geom, ST_Union(b.{CAR_GEOM})), 97823)) / 
     NULLIF(ST_Area(ST_Transform(a.geom, 97823)), 0)) AS perc_area_sobreposta,
    -- geometria da interseção (para visualização)
    ST_Intersection(a.geom, ST_Union(b.{CAR_GEOM})) AS geom_intersecao

FROM car_interesse a
CROSS JOIN car_sobrepostos_unicos b
GROUP BY a.geom;
"""

# ============================================================================
# Exibição da query
# ============================================================================

md_query_perc_area = f"""
<div style="background-color:#f5f5f5; padding:15px; border-radius:8px; border-left: 5px solid #2c6e9e; font-family: monospace; font-size: 0.9em;">
<b> Consulta SQL executada:</b><br>
<code>
{query_perc_area_sobreposta_total}
</code>
</div>
"""

display(Markdown(md_query_perc_area))
```

```python
# ============================================================================
# Executa a consulta e carrega os resultados
# ============================================================================

df_perc_area_sobreposta = pd.read_sql(text(query_perc_area_sobreposta_total), con=engine)

# Extrai os valores
area_imovel_ha = df_perc_area_sobreposta['area_imovel_ha'].iloc[0]
area_sobreposta_ha = df_perc_area_sobreposta['area_sobreposta_ha'].iloc[0]
perc_area_sobreposta = df_perc_area_sobreposta['perc_area_sobreposta'].iloc[0] * 100

# ============================================================================
# Exibe o resultado da métrica B
# ============================================================================
def formatar_br(valor):
    return f"{valor:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")

md_metrica_b = f"""
<div style="background-color:#e8f4fd; padding:15px; border-radius:8px; border-left: 5px solid #2196f3; margin: 15px 0;">
<b> Métrica B - Percentual de sobreposição</b><br>
<br>
<b>Área do imóvel (CAR de interesse):</b> {formatar_br(area_imovel_ha)} ha<br>
<b>Área sobreposta total:</b> {formatar_br(area_sobreposta_ha)} ha<br>
<b>Percentual sobreposto:</b> {formatar_br(perc_area_sobreposta)}%
</div>
"""

display(Markdown(md_metrica_b))

# ============================================================================
# Cria GeoDataFrame com a geometria da interseção (se existir)
# ============================================================================

from shapely import wkb
import binascii

geom_intersecao_raw = df_perc_area_sobreposta['geom_intersecao'].iloc[0]

if geom_intersecao_raw is not None:
    try:
        # Converte de hexadecimal WKB para bytes
        if isinstance(geom_intersecao_raw, str):
            geom_bytes = binascii.unhexlify(geom_intersecao_raw)
        else:
            geom_bytes = geom_intersecao_raw
        
        # Carrega a geometria a partir dos bytes WKB
        geom_intersecao = wkb.loads(geom_bytes)
        
        if not geom_intersecao.is_empty:
            gdf_intersecao = gpd.GeoDataFrame(
                {'geometry': [geom_intersecao]}, 
                crs=4674
            )
            print("Geometria da interseção carregada com sucesso.")
        else:
            gdf_intersecao = None
            print("A geometria da interseção está vazia.")
    except Exception as e:
        gdf_intersecao = None
        print(f"Não foi possível carregar a geometria da interseção: {e}")
else:
    gdf_intersecao = None
    print("Não há interseção entre o imóvel e os CARs sobrepostos.")
```

### 4.2.5: Identificar sobreposição com CAR semelhante ao SIGEF (Métrica C)

A Métrica C verifica se há sobreposição entre o CAR de interesse e **outros imóveis do CAR que apresentam semelhança com o SIGEF**, conforme o indicador desenvolvido anteriormente.

A presença desse tipo de sobreposição é relevante, pois o SIGEF é uma base associada a processos formais de certificação fundiária, com maior rigor na delimitação dos imóveis.


**Como interpretar**

- Caso **não haja sobreposição com CARs semelhantes ao SIGEF**:
  - a sobreposição pode estar associada a inconsistências entre cadastros do próprio CAR;

- Caso **haja sobreposição com CARs semelhantes ao SIGEF**:
  - pode indicar maior probabilidade de inconsistência no CAR de interesse;

**Síntese**

A Métrica C complementa as métricas de intensidade, permitindo distinguir entre:
- sobreposições mais simples (entre cadastros com menor evidência fundiária);
- e sobreposições potencialmente mais críticas (envolvendo imóveis com correspondência ao SIGEF).

#### Identificar CARs sobrepostos que são semelhantes ao SIGEF (Métrica C)


<div style="background-color:#e8f4fd; padding:15px; border-radius:8px; border-left: 5px solid #2196f3; margin: 15px 0;">
<b>Definindo o limiar</b><br>
O limiar define a partir de quais valores de p1 e p2 um imóvel será considerado semelhante ao SIGEF (Aula 7).<br>
Este valor pode ser ajustado conforme o nível de rigor desejado na análise.
</div>

```python
limiar = 0.9
```

```python
# ============================================================================
# Query para identificar CARs sobrepostos com semelhante ao SIGEF (métrica C)
# ============================================================================
query_match_sigef = f"""
WITH car_interesse AS (
    SELECT DISTINCT ON ({CAR_ID})
        {CAR_ID},
        {CAR_GEOM} AS geom
    FROM {CAR_SCHEMA}.{CAR_TABLE}
    WHERE {CAR_ID} = '{COD_IMOVEL}'
      AND {CAR_STATUS} <> 'CA' AND {CAR_CONDICAO} NOT ILIKE '%cancelado%'
    ORDER BY 
        {CAR_ID}, 
        COALESCE({CAR_DATA_ATUALIZACAO}, {CAR_DATA_CRIACAO}) DESC
),

car_sobrepostos AS (
    SELECT 
        b.*
    FROM car_interesse a
    INNER JOIN {CAR_SCHEMA}.{CAR_TABLE} b
        ON ST_Intersects(a.geom, b.{CAR_GEOM})
        AND ST_Area(ST_Transform(ST_Intersection(a.geom, b.{CAR_GEOM}),97823)) > 20
    WHERE 
        b.{CAR_ID} <> a.{CAR_ID}
        AND b.{CAR_STATUS} <> 'CA' AND b.{CAR_CONDICAO} NOT ILIKE '%cancelado%'
),

car_sobrepostos_unicos AS (
    SELECT DISTINCT ON ({CAR_ID})
        {CAR_ID},
        {CAR_GEOM} AS geom
    FROM car_sobrepostos
    ORDER BY 
        {CAR_ID},
        COALESCE({CAR_DATA_ATUALIZACAO}, {CAR_DATA_CRIACAO}) DESC
),

sigef_intersect AS (
    SELECT
        c.{CAR_ID},
        s.codigo_imo,
        s.parcela_co,
        s.nome_area,

        ST_Area(ST_Transform(c.geom, 97823)) / 10000 AS area_car,
        ST_Area(ST_Transform(s.{SIGEF_GEOM}, 97823)) / 10000 AS area_sigef_ha,

        ST_Area(
            ST_Transform(
                ST_Intersection(s.{SIGEF_GEOM}, c.geom),
                97823
            )
        ) / 10000 AS area_intersecao_ha

    FROM car_sobrepostos_unicos c
    INNER JOIN {SIGEF_SCHEMA}.{SIGEF_TABLE} s
        ON ST_Intersects(s.{SIGEF_GEOM}, c.geom)
),

calculo_parcela AS (
    SELECT
        {CAR_ID},
        codigo_imo,
        'por_parcela' AS tipo,

        area_intersecao_ha / NULLIF(area_car, 0) AS p1,
        area_intersecao_ha / NULLIF(area_sigef_ha, 0) AS p2
    FROM sigef_intersect
),

calculo_agrupado AS (
    SELECT
        {CAR_ID},
        codigo_imo,
        'agrupamento' AS tipo,

        SUM(area_intersecao_ha) / NULLIF(MAX(area_car), 0) AS p1,
        SUM(area_intersecao_ha) / NULLIF(SUM(area_sigef_ha), 0) AS p2

    FROM sigef_intersect
    GROUP BY {CAR_ID}, codigo_imo
),

indicador_final AS (
    SELECT *,
        CASE
            WHEN p1 > {limiar} AND p2 > {limiar} THEN TRUE
            ELSE FALSE
        END AS semelhante_sigef
    FROM calculo_parcela

    UNION ALL

    SELECT *,
        CASE
            WHEN p1 > {limiar} AND p2 > {limiar} THEN TRUE
            ELSE FALSE
        END AS semelhante_sigef
    FROM calculo_agrupado
)

SELECT DISTINCT
    {CAR_ID},
    codigo_imo,
    semelhante_sigef
FROM indicador_final
WHERE semelhante_sigef
ORDER BY {CAR_ID}, codigo_imo;
"""

# ============================================================================
# Exibição da query
# ============================================================================

md_query_match = f"""
<div style="background-color:#f5f5f5; padding:15px; border-radius:8px; border-left: 5px solid #2c6e9e; font-family: monospace; font-size: 0.9em;">
<b> Consulta SQL executada:</b><br>
<code>
{query_match_sigef.replace('<', '&lt;').replace('>', '&gt;')}
</code>
</div>
"""

display(Markdown(md_query_match))
```

```python
# ============================================================================
# Executa a consulta e carrega os resultados
# ============================================================================

df_match_sigef = pd.read_sql(text(query_match_sigef), con=engine)

# Verifica se há dados
if df_match_sigef.empty:
    n_match = 0
    n_total = 0
    codigos_match_completos = []
else:
    # Conta CARs únicos com match
    n_match = df_match_sigef[
        df_match_sigef['semelhante_sigef']
    ]['cod_imovel'].nunique()

    # Total de CARs avaliados
    n_total = df_match_sigef['cod_imovel'].nunique()

    # Lista de códigos
    codigos_match_completos = df_match_sigef[
        df_match_sigef['semelhante_sigef']
    ]['cod_imovel'].unique().tolist()


# ============================================================================
# Exibe o resultado da métrica D (usando apenas Markdown, sem HTML cru)
# ============================================================================

if n_match > 0:
    # Lista os códigos em formato de lista Markdown
    lista_codigos = ""
    for cod in codigos_match_completos:
        lista_codigos += f"- `{cod}`\n"
    
    resultado_texto = f"""
**CARs sobrepostos com match SIGEF:** {n_match}

**Códigos dos imóveis com match:**

{lista_codigos}
"""
    cor_borda = "green"
else:
    resultado_texto = f"""
**Nenhum CAR sobreposto possui match com SIGEF.**

Total de CARs sobrepostos: {n_match} de {n_total}
"""
    cor_borda = "orange"

# Usa Markdown simples
display(Markdown(f"""
<div style="background-color:#f0f7ff; padding:15px; border-radius:8px; border-left: 5px solid {cor_borda}; margin: 15px 0;">
<b> Métrica D - Sobreposição com CAR que possui match SIGEF</b>
<br><br>
{resultado_texto}
</div>
"""))

display(df_match_sigef)
```

```python
## Metricas A e B
query_num_perc_sobreposicao = f"""
WITH car_interesse AS (
    SELECT DISTINCT ON ({CAR_ID})
        {CAR_ID},
        {CAR_GEOM} AS geom
    FROM {CAR_SCHEMA}.{CAR_TABLE}
    WHERE {CAR_ID} = '{COD_IMOVEL}'
        AND {CAR_STATUS} <> 'CA' AND {CAR_CONDICAO} NOT ILIKE '%cancelado%'
    ORDER BY 
        {CAR_ID}, 
        COALESCE({CAR_DATA_ATUALIZACAO}, {CAR_DATA_CRIACAO}) DESC
),
car_sobrepostos AS (
    SELECT 
        b.*
    FROM car_interesse a
    INNER JOIN {CAR_SCHEMA}.{CAR_TABLE} b
        ON ST_Intersects(a.geom, b.{CAR_GEOM})
        AND ST_Area(ST_Transform(ST_Intersection(a.geom, b.{CAR_GEOM}),97823)) > 0
    WHERE 
        b.{CAR_ID} <> a.{CAR_ID}
        AND {CAR_STATUS} <> 'CA' AND {CAR_CONDICAO} NOT ILIKE '%cancelado%'
),
car_sobrepostos_unicos AS (
SELECT DISTINCT ON ({CAR_ID})
*
FROM car_sobrepostos
ORDER BY 
    {CAR_ID},
    COALESCE({CAR_DATA_ATUALIZACAO}, {CAR_DATA_CRIACAO}) DESC
)
SELECT
    -- número de imóveis sobrepostos
    count(b.{CAR_ID}) AS num_sobrepostos,
    -- área do imóvel em hectares
    ST_Area(ST_Transform(a.geom, 97823)) / 10000 AS area_imovel_ha,
    -- área total sobreposta em hectares (união de todos os CARs sobrepostos)
    ST_Area(ST_Transform(ST_Intersection(a.geom, ST_Union(b.{CAR_GEOM})), 97823)) / 10000 AS area_sobreposta_ha,
    -- percentual da área do imóvel que está sobreposta
    (ST_Area(ST_Transform(ST_Intersection(a.geom, ST_Union(b.{CAR_GEOM})), 97823)) / 
     NULLIF(ST_Area(ST_Transform(a.geom, 97823)), 0)) AS perc_area_sobreposta

FROM car_interesse a
CROSS JOIN car_sobrepostos_unicos b
GROUP BY a.geom ;
"""

## Metrica C
query_match_sigef = f"""
WITH car_interesse AS (
    SELECT DISTINCT ON ({CAR_ID})
        {CAR_ID},
        {CAR_GEOM} AS geom
    FROM {CAR_SCHEMA}.{CAR_TABLE}
    WHERE {CAR_ID} = '{COD_IMOVEL}'
      AND {CAR_STATUS} <> 'CA' AND {CAR_CONDICAO} NOT ILIKE '%cancelado%'
    ORDER BY 
        {CAR_ID}, 
        COALESCE({CAR_DATA_ATUALIZACAO}, {CAR_DATA_CRIACAO}) DESC
),

car_sobrepostos AS (
    SELECT 
        b.*
    FROM car_interesse a
    INNER JOIN {CAR_SCHEMA}.{CAR_TABLE} b
        ON ST_Intersects(a.geom, b.{CAR_GEOM})
        AND ST_Area(ST_Transform(ST_Intersection(a.geom, b.{CAR_GEOM}),97823)) > 20
    WHERE 
        b.{CAR_ID} <> a.{CAR_ID}
        AND b.{CAR_STATUS} <> 'CA' AND b.{CAR_CONDICAO} NOT ILIKE '%cancelado%'
),

car_sobrepostos_unicos AS (
    SELECT DISTINCT ON ({CAR_ID})
        {CAR_ID},
        {CAR_GEOM} AS geom
    FROM car_sobrepostos
    ORDER BY 
        {CAR_ID},
        COALESCE({CAR_DATA_ATUALIZACAO}, {CAR_DATA_CRIACAO}) DESC
),

sigef_intersect AS (
    SELECT
        c.{CAR_ID},
        s.codigo_imo,
        s.parcela_co,
        s.nome_area,

        ST_Area(ST_Transform(c.geom, 97823)) / 10000 AS area_car,
        ST_Area(ST_Transform(s.{SIGEF_GEOM}, 97823)) / 10000 AS area_sigef_ha,
        ST_Area(ST_Transform(ST_Intersection(s.{SIGEF_GEOM}, c.geom),97823)) / 10000 AS area_intersecao_ha

    FROM car_sobrepostos_unicos c
    INNER JOIN {SIGEF_SCHEMA}.{SIGEF_TABLE} s
        ON ST_Intersects(s.{SIGEF_GEOM}, c.geom)
),

calculo_parcela AS (
    SELECT
        {CAR_ID},
        codigo_imo,
        'por_parcela' AS tipo,

        area_intersecao_ha / NULLIF(area_car, 0) AS p1,
        area_intersecao_ha / NULLIF(area_sigef_ha, 0) AS p2
    FROM sigef_intersect
),

calculo_agrupado AS (
    SELECT
        {CAR_ID},
        codigo_imo,
        'agrupamento' AS tipo,

        SUM(area_intersecao_ha) / NULLIF(MAX(area_car), 0) AS p1,
        SUM(area_intersecao_ha) / NULLIF(SUM(area_sigef_ha), 0) AS p2

    FROM sigef_intersect
    GROUP BY {CAR_ID}, codigo_imo
),

indicador_final AS (
    SELECT *,
        CASE
            WHEN p1 > {limiar} AND p2 > {limiar} THEN TRUE
            ELSE FALSE
        END AS semelhante_sigef
    FROM calculo_parcela

    UNION ALL

    SELECT *,
        CASE
            WHEN p1 > {limiar} AND p2 > {limiar} THEN TRUE
            ELSE FALSE
        END AS semelhante_sigef
    FROM calculo_agrupado
)

SELECT DISTINCT
    {CAR_ID},
    codigo_imo,
    semelhante_sigef
FROM indicador_final
WHERE semelhante_sigef
ORDER BY {CAR_ID}, codigo_imo;
"""


# ============================================================================
# Exibição da query
# ============================================================================

md_query_num_perc = f"""
<div style="background-color:#f5f5f5; padding:15px; border-radius:8px; border-left: 5px solid #2c6e9e; font-family: monospace; font-size: 0.9em;">
<b> Consulta SQL executada (Métricas A e B):</b><br>
<code>
{query_num_perc_sobreposicao}
</code>
</div>
"""

display(Markdown(md_query_num_perc))


md_query_match_sigef = f"""
<div style="background-color:#f5f5f5; padding:15px; border-radius:8px; border-left: 5px solid #2c6e9e; font-family: monospace; font-size: 0.9em;">
<b> Consulta SQL executada (Métrica C):</b><br>
<code>
{query_match_sigef}
</code>
</div>
"""

display(Markdown(md_query_match_sigef))
```

```python
# ============================================================================
# Executa a consulta e carrega os resultados
# ============================================================================
# Métrica A e B
df_query_num_perc_sobreposicao = pd.read_sql(text(query_num_perc_sobreposicao), con=engine)

# Extrai os valores
n_sobreposicoes = df_query_num_perc_sobreposicao['num_sobrepostos'].iloc[0] 
area_imovel_ha = df_query_num_perc_sobreposicao['area_imovel_ha'].iloc[0]
area_sobreposta_ha = df_query_num_perc_sobreposicao['area_sobreposta_ha'].iloc[0]
perc_area_sobreposta = df_query_num_perc_sobreposicao['perc_area_sobreposta'].iloc[0] * 100

#Métrica C
df_match_sigef_final = pd.read_sql(text(query_match_sigef), con=engine)

# Verifica se há dados
if df_match_sigef_final.empty:
    n_match = 0
    n_total = 0
    codigos_match_completos = []
else:
    # Conta CARs únicos com match
    n_match = df_match_sigef_final[
        df_match_sigef_final['semelhante_sigef']
    ]['cod_imovel'].nunique()

    # Total de CARs avaliados
    n_total = df_match_sigef_final['cod_imovel'].nunique()

    # Lista de códigos
    codigos_match_completos = df_match_sigef_final[
        df_match_sigef_final['semelhante_sigef']
    ]['cod_imovel'].unique().tolist()

# ============================================================================
# Exibe o resultado da métrica A e B
# ============================================================================
def formatar_br(valor):
    return f"{valor:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")

md_metrica_ab = f"""
<div style="background-color:#e8f4fd; padding:15px; border-radius:8px; border-left: 5px solid #2196f3; margin: 15px 0;">
<b> Dimensão de Intensidade (Métricas A e B) </b><br>
<br>
<b>Número de imóveis sobrepostos:</b> {n_sobreposicoes}<br>
<b>Área do imóvel (CAR de interesse):</b> {formatar_br(area_imovel_ha)} ha<br>
<b>Área sobreposta total:</b> {formatar_br(area_sobreposta_ha)} ha<br>
<b>Percentual sobreposto:</b> {formatar_br(perc_area_sobreposta)}%
</div>
"""

display(Markdown(md_metrica_ab))


# ============================================================================
# Exibe o resultado da métrica C 
# ============================================================================

if n_match > 0:
    # Lista os códigos em formato de lista Markdown
    lista_codigos = ""
    for cod in codigos_match_completos:
        lista_codigos += f"- `{cod}`\n"
    
    resultado_texto = f"""
**Número de CARs sobrepostos ao CAR de interesse que são semelhantes ao SIGEF:** {n_match}

**Códigos dos imóveis com match:**

{lista_codigos}
"""
    cor_borda = "green"
else:
    resultado_texto = f"""
**Nenhum CAR sobreposto possui match com SIGEF.**

Total de CARs sobrepostos: {n_match} de {n_total}
"""
    cor_borda = "orange"

# Usa Markdown simples
display(Markdown(f"""
<div style="background-color:#f0f7ff; padding:15px; border-radius:8px; border-left: 5px solid {cor_borda}; margin: 15px 0;">
<b> Dimensão de Qualidade (Métrica C)</b>
<br><br>
{resultado_texto}
</div>
"""))

display(df_match_sigef_final)
```
